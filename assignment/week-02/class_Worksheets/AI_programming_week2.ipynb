```python
# Step 5: Basic Data Cleaning

# Create a copy of the original dataset before cleaning
# This helps us keep the original data unchanged
df_clean = df.copy()

# Display the number of missing values in each column
# This helps us identify which columns have empty values
print("Missing values before cleaning:")
print(df_clean.isnull().sum())

# Display the number of duplicate rows in the dataset
# Duplicate rows may affect the model training process
print("\nNumber of duplicate rows before cleaning:")
print(df_clean.duplicated().sum())

# Display the data type of each column
# This helps us check whether columns are stored in the correct format
print("\nData types before cleaning:")
print(df_clean.dtypes)
```

```python
# Fill missing values in the Income column with the median income
# The median is often used because it is less affected by very large or very small values
df_clean["Income"] = df_clean["Income"].fillna(df_clean["Income"].median())

# Fill missing values in the City column with the most frequent city
# mode()[0] returns the most common value in the column
df_clean["City"] = df_clean["City"].fillna(df_clean["City"].mode()[0])

# Fill missing values in the Membership_Level column with the most frequent membership level
# This is useful for categorical columns
df_clean["Membership_Level"] = df_clean["Membership_Level"].fillna(df_clean["Membership_Level"].mode()[0])

# Remove duplicate rows from the dataset
# This keeps only one copy of repeated records
df_clean = df_clean.drop_duplicates()

# Display the dataset after removing duplicates
df_clean.head()
```

```python
# Convert Customer_ID to integer
# Customer_ID is an identifier, not a prediction feature
df_clean["Customer_ID"] = df_clean["Customer_ID"].astype(int)

# Convert Age to integer
# Age should be stored as a numeric whole number
df_clean["Age"] = df_clean["Age"].astype(int)

# Convert Income to float
# Income is numeric and may contain decimal values after cleaning
df_clean["Income"] = df_clean["Income"].astype(float)

# Convert Website_Visits to integer
# Website visits should be numeric
df_clean["Website_Visits"] = df_clean["Website_Visits"].astype(int)

# Convert Previous_Purchases to integer
# Previous purchases should be numeric
df_clean["Previous_Purchases"] = df_clean["Previous_Purchases"].astype(int)

# Convert categorical columns to category type
# This helps us clearly separate categorical variables from numerical variables
df_clean["Gender"] = df_clean["Gender"].astype("category")
df_clean["City"] = df_clean["City"].astype("category")
df_clean["Membership_Level"] = df_clean["Membership_Level"].astype("category")
df_clean["Purchased"] = df_clean["Purchased"].astype("category")

# Display data types after cleaning
print("Data types after cleaning:")
print(df_clean.dtypes)
```

```python
# Check missing values again after cleaning
print("Missing values after cleaning:")
print(df_clean.isnull().sum())

# Check duplicate rows again after cleaning
print("\nNumber of duplicate rows after cleaning:")
print(df_clean.duplicated().sum())

# Display the shape of the cleaned dataset
print("\nShape of cleaned dataset:")
print(df_clean.shape)

# Display the first 5 rows of the cleaned dataset
df_clean.head()
```
